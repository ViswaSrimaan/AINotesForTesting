A processing unit, also known as a central processing unit (CPU) or processor, is a complex set of electronic circuitry that's the core computational unit in a computer or server:
Function
The CPU is responsible for processing data and making computing possible. It interprets, processes, and executes instructions from hardware and software programs, performing arithmetic, logic, and other operations to transform data input into more usable information output. The CPU also guides the computer through the various steps of solving a problem, handling all computing tasks required for running the operating system and applications

### Comparison of TPU, GPU, and CPU in AI

| Feature              | TPU (Tensor Processing Unit)                                  | GPU (Graphics Processing Unit)                                 | CPU (Central Processing Unit)                                    |
|----------------------|---------------------------------------------------------------|----------------------------------------------------------------|-----------------------------------------------------------------|
| **Purpose**          | Designed specifically for accelerating machine learning tasks | Optimized for parallel processing, originally for graphics     | General-purpose processing, optimized for sequential tasks      |
| **Architecture**     | Application-specific integrated circuit (ASIC) for ML tasks   | Thousands of smaller cores for parallel tasks                  | Few cores optimized for single-threaded performance             |
| **Key Strength**     | High efficiency for TensorFlow-based neural network tasks     | High throughput for parallel tasks                             | Versatility in handling a wide range of tasks                   |
| **Performance**      | Exceptional for deep learning with TensorFlow                 | Excellent for deep learning, computer vision, and NLP          | Adequate for smaller AI models, data preprocessing              |
| **Memory Bandwidth** | High, optimized for ML workloads                              | High, suitable for large data transfers                        | Moderate, suitable for general computing tasks                  |
| **Memory Management**| Explicitly managed                                            | Mixed                                                          | Implicitly managed                                              |
| **Memory Hierarchy** | Activation Buffer, Weight FIFO, Accumulation Register File    | L2, Shared Memory (SM), Register File (RF), L1/Texture (L1/TX) | L1 Cache (L1D, L1I), L2 Cache, L3 Cache                         |
| **Compute Primitive**| Tensor                                                        | Vector                                                         | Scalar                                                          |
| **Power Efficiency** | More energy-efficient for specific AI tasks                   | Higher power consumption due to parallel processing capabilities| Generally more power-efficient for single-thread tasks          |
| **Cost**             | Generally available through cloud services (e.g., Google Cloud)| Typically more expensive than CPUs                             | Typically less expensive, widely available                      |
| **Availability**     | Primarily available through Google Cloud                      | Widely available in consumer and professional markets          | Ubiquitous, found in all types of computing devices             |
| **Ease of Use**      | Requires setup via cloud platforms, TensorFlow integration    | Requires proper driver and library support (e.g., CUDA for NVIDIA)| Easy to use, no special setup needed for basic tasks            |
| **Ideal Use Cases**  | Training large, complex models, inference with TensorFlow     | Training deep neural networks, real-time data processing       | Data preprocessing, smaller AI models, general-purpose tasks    |
| **Examples**         | TensorFlow models, Google Cloud ML services                   | PyTorch, TensorFlow models, computer vision applications       | Data analysis scripts, smaller machine learning models          |
| **Training Speed**   | Fast for supported frameworks                                 | Fast for parallelizable tasks                                  | Slower compared to TPU and GPU for large-scale AI tasks         |
| **Parallelism**      | High, optimized for parallel processing in ML tasks           | Very high, handles thousands of parallel threads               | Limited parallelism compared to GPU and TPU                     |


# CPU - By default, Google Colab uses the CPU for computations.

```python
# Import Libraries
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load and Prepare the Data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
y_train, y_test = to_categorical(y_train), to_categorical(y_test)

# Define the Model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the Model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the Model
model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)

# Evaluate the Model
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print('\nTest accuracy:', test_acc)
```

# TPU - To use a TPU in Google Colab, you need to enable it and configure TensorFlow to use the TPU.
Follow these steps:

Enable TPU:

Go to Runtime > Change runtime type > Select TPU > Save.

```python
# Connect to the TPU
import tensorflow as tf

# Detect and initialize the TPU
try:
    tpu = tf.distribute.cluster_resolver.TPUClusterResolver()  # TPU detection
    tf.config.experimental_connect_to_cluster(tpu)
    tf.tpu.experimental.initialize_tpu_system(tpu)
    strategy = tf.distribute.TPUStrategy(tpu)
    print('Running on TPU ', tpu.cluster_spec().as_dict()['worker'])
except ValueError:
    strategy = tf.distribute.get_strategy()  # Default strategy if no TPU is detected
    print('Running on default strategy')

# Import Libraries
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load and Prepare the Data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
y_train, y_test = to_categorical(y_train), to_categorical(y_test)

# Define and Compile the Model within the TPU Strategy Scope
with strategy.scope():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

# Train the Model
model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)

# Evaluate the Model
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print('\nTest accuracy:', test_acc)
```
