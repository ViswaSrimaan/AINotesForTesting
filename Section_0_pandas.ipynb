### 4. Creating Data Structures

#### **Creating a Series**

A Series is a one-dimensional array-like object containing an array of data and an associated array of data labels, called its index.

- **From a List**:
  ```python
  import pandas as pd

  data = [1, 2, 3, 4, 5]
  series = pd.Series(data)
  print(series)
  ```

- **From a Dictionary**:
  ```python
  data = {'a': 1, 'b': 2, 'c': 3}
  series = pd.Series(data)
  print(series)
  ```


````python
import pandas as pd

data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print(series)
````

````python
data = {'a': 1, 'b': 2, 'c': 3}
series = pd.Series(data)
print(series)
````

#### **Creating a DataFrame**

A DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

- **From a Dictionary**:

  ```python

  print(df)
  ```


### 8. Merging and Joining DataFrames

- **Merging**:
  ```python
  df1 = pd.DataFrame({
      'Name': ['John', 'Anna', 'Peter'],
      'Age': [28, 24, 35]
  })
  df2 = pd.DataFrame({
      'Name': ['John', 'Anna', 'Peter'],
      'City': ['New York', 'Paris', 'Berlin']
  })
  merged_df = pd.merge(df1, df2, on='Name')
  print(merged_df)
  ```

- **Joining**:
  ```python
  left = pd.DataFrame({
      'key': ['K0', 'K1', 'K2'],
      'A': ['A0', 'A1', 'A2']
  })
  right = pd.DataFrame({
      'key': ['K0', 'K1', 'K2'],
      'B': ['B0', 'B1', 'B2']
  })
  joined_df = left.join(right.set_index('key'), on='key')
  print(joined_df)
  ```
---

````python
left = pd.DataFrame({
    'key': ['K0', 'K1', 'K2'],
    'A': ['A0', 'A1', 'A2']
})
right = pd.DataFrame({
    'key': ['K0', 'K1', 'K2'],
    'B': ['B0', 'B1', 'B2']
})
joined_df = left.join(right.set_index('key'), on='key')
print(joined_df)
````

````python
df1 = pd.DataFrame({
    'Name': ['John', 'Anna', 'Peter'],
    'Age': [28, 24, 35]
})
df2 = pd.DataFrame({
    'Name': ['John', 'Anna', 'Peter'],
    'City': ['New York', 'Paris', 'Berlin']
})
merged_df = pd.merge(df1, df2, on='Name')
print(merged_df)
````


### 9. Advanced Operations

#### **Pivot Tables**

- **Creating Pivot Tables**:
  ```python
  pivot_table = df.pivot_table(values='Age', index='City', columns='Name', aggfunc='mean')
  print(pivot_table)
  ```

#### **Time Series Analysis**

- **Working with Dates**:
  ```python
  df['Date'] = pd.to_datetime(df['Date'])
  df.set_index('Date', inplace=True)
  print(df)
  ```

- **Resampling**:
  ```python
  df = df.resample('M').mean()
  print(df)
  ```


````python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
print(df)
````
